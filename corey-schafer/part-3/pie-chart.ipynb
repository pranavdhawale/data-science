```python
import matplotlib.pyplot as plt

plt.style.use("fivethirtyeight")
```

```python
slices = [67,33]

plt.pie(slices)

plt.show()
```

```python
slices = [67, 33, 75, 54, 98, 46]
labels = ['a', 'b', 'c', 'd', 'e', 'f']

plt.pie(slices, labels=labels, 
        wedgeprops={'edgecolor' : 'black'})

plt.title("Awesome Pie Chart")

plt.show()
```
