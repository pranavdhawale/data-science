```python
import matplotlib.pyplot as plt

plt.style.use("seaborn")
```

```python
slices = [67,33]

plt.pie(slices)

plt.show()
```

```python
slices = [67, 33, 75, 54, 98, 46]
labels = ['a', 'b', 'c', 'd', 'e', 'f']

plt.pie(slices, labels=labels, 
        wedgeprops={'edgecolor' : 'black'})

plt.title("Awesome Pie Chart")

plt.show()
```

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 'C', 
        'Other(s):', 'Ruby', 'Go', 'Assembly']
colors = ["#F7DF1E", "#E34F26", "#4479A1", "#3776AB", "#FC4C02", "#4EAA25", "#280068", "#777BB4", "#00599C", "#3178C6", "#A8B9CC", 
        "#4D2B1A", "#CC342D", "#00ADD8", "#007AAC"]
```

```python
plt.pie(slices, labels=labels, colors=colors, 
        wedgeprops={'edgecolor' : 'black'})

plt.title("Programming Language Priority")

plt.show()
```

```python
plt.pie(slices[0:5], labels=labels[0:5], colors=colors[0:5], 
        wedgeprops={'edgecolor' : 'black'})

plt.title("Programming Language Priority")

plt.show()
```
