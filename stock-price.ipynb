```python
import cv2
import matplotlib.pyplot as plt
```

```python
apl_price = [93.96, 112.5, 104.72, 143.63, 169.6, 102.12]
ms_price = [39.05, 50.26, 120.72, 69.98, 53.2, 75.2]
year = [2010, 2011, 2012, 2013, 2014, 2015]
```

```python
# One single line to show graph in one plotting
plt.plot(year,apl_price,
        year, ms_price)

# Scatter Plotting

plt.scatter(year,apl_price)
plt.scatter(year,ms_price)

# Label the X and Y axis
plt.xlabel("Year")
plt.ylabel("Price")

plt.show()
```

```python
fig_1 = plt.figure(1, figsize=(20, 4))

plot_price = fig_1.add_subplot(121)
scatter_price = fig_1.add_subplot(122)

plot_price.plot(year, apl_price,
            year, ms_price)
scatter_price.scatter(year, apl_price)
scatter_price.scatter(year, ms_price)
```
